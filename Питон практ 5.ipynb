Функция nunique() возвращает количество уникальных значений в каждом столбце объекта. В столбце 'gender' ввсего 2 уникальных значений, в столбце 'race/ethnicity' 5 уникальных значений, в 'parental level of education' - 6 уникальных значений, в 'lunch' - 2 уникальных значений, в 'test preparation course' - 2 уникальных значений, в 'math score' - 81 уникальных значений, в 'reading score' - 72 уникальных значений, в 'writing score' - 77 уникальных значений.

Код df.sortvalues(by='math score') используется для сортировки датасета по заданному столбцу 'math score'. Это означает, что строки в датасете будут упорядочены в порядке возрастания значений в столбце 'math score', то есть строки с более низкими значеними 'math score' будут отображаться в начале, а строки с более высокими значениями будут отображаться в конце после сортировки. 

```python
df.sort_values(by='math score')
```

Код df.drop(columns="race/ethnicity") используется для удаления столбца "race/ethnicity" из датасета df. При вызове этой функции, заданный столбец "race/ethnicity" будет удалён из датасет. Датасет будет изменен и больше не будет содержать столбец "race/ethnicity".

```python
df.drop(columns=["race/ethnicity"])
```

Этот код используется для замены значений в столбце 'parental level of education' в датасете. Значения 'some college' в столбце 'parental level of education' будут заменены на 'Not specified'. Таким образом, после выполнения этой операции, все строки в столбце 'parental level of education', которые ранее содержали значение 'some college', будут теперь содержать значение 'Not specified'.

```python
df['parental level of education'] = df['parental level of education'].replace('some college', 'Not specified')
```

Эта строка удаляет дубликаты в столбце 'writing score' в датасете.

```python
df.drop_duplicates(subset=['writing score'])
```
